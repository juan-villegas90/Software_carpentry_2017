```python
def mean_of_column (state_name, column_name):
    state_df = health_data.loc[health_data['State'] == state_name,[column_name]]
    return state_df[column_name].mean()
```

```python
mean_of_column('AK','Total_Births')
```

```python
uniques = health_data['State'].unique()
print(uniques)
type(uniques)
```

```python
list(health_data['State'].unique())
```

```python
bar_chart = pygal.Bar()
bar_chart.title = 'Mean of Poverty across states'
for state in list(health_data['State'].unique()):
    bar_chart.add(state,mean_of_column(state,'Total_Births'))
SVG(bar_chart.render())
```
